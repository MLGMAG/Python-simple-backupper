```python
import json

def get_article_data(input_filepath):
    json_file = open(input_filepath)
    return json.load(json_file)

def write_json(data, output_filepath):
    with open(output_filepath, 'w', encoding='utf8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
import spacy

nlp_uk = spacy.load("uk_core_news_md")

def get_ner_spacy(text):
    doc = nlp_uk(text)

    result_set = set()
    for ent in doc.ents:
        obj = {"lemma": ent.lemma_, "label": ent.label_}
        result_set.add(frozenset(obj.items()))

    result = []
    for entity in result_set:
        entity_dict = dict(entity)
        entity_lemma = entity_dict['lemma']
        entity_label = entity_dict['label']
        result.append([entity_lemma, entity_label])

    result.sort()
    return result
```

```python
def get_articles_ner(articles):
    articles_ner = []    
    
    for article in articles:
        article_ner = {
            'article': article['article'],
            'link': article['link'],
            'header': article['header'],
            'text': article['text'],
            'header_ner': get_ner_spacy(article['header']),
            'text_ner': get_ner_spacy(article['text']),
        }
        articles_ner.append(article_ner)
    
    return articles_ner
```

```python
INPUT_FILE_PATH = './resources/input_articles.json'
OUTPUT_FILE_PATH = './resources/output_articles.json'

articles = get_article_data(INPUT_FILE_PATH)
articles_ner = get_articles_ner(articles)
write_json(articles_ner, OUTPUT_FILE_PATH)
```
