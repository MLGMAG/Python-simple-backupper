```python
import pandas as pd

REVIEWS_FILE = 'reviews.jsonlines'

reviews = pd.read_json(REVIEWS_FILE, lines=True)

book_id_rate = reviews.groupby('book_id').agg({'rating': 'mean'}).round(2)
book_id_rate
```

```python
BOOKS_FILE = 'books.jsonlines'

books = pd.read_json(BOOKS_FILE, lines=True, nrows=1000)
book_id_book = books.set_index('book_id')
book_id_book.update(book_id_rate)
books_with_updated_rating = book_id_book.reset_index()
```
